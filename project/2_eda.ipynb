```python
import os, json
from collections import Counter

DATA_DIR = "c:\\Users\\main\\Desktop\\train_labeled"
json_files = [f for f in os.listdir(DATA_DIR) if f.endswith(".json")]

class_cnt = Counter()
res_cnt = Counter()
mixed_cnt = 0

for fname in json_files:
    path = os.path.join(DATA_DIR, fname)
    obj = json.load(open(path, "r", encoding="utf-8"))

    # 1) 해상도 집계 (images 아래)
    w = obj["images"]["image_width"]
    h = obj["images"]["image_height"]
    res_cnt[f"{w}×{h}"] += 1

    # 2) 클래스 집계 (annotations 리스트 아래)
    for ann in obj["annotations"]:
        class_cnt[ann["class_name"]] += 1
        if ann["class_name"] == "mixed":
            mixed_cnt += 1
            print(fname)

print(f"mixed 갯수 = {mixed_cnt}")

total_images = len(json_files)
total_annotations = sum(class_cnt.values())

# 결과를 노션용 Markdown 테이블로 출력
print("#### 클래스 분포\n")
print("| 클래스       | 어노테이션 수 | 비율    |")
print("|---          |---:          |---:     |")
for cls, cnt in class_cnt.most_common():
    pct = cnt / total_annotations * 100
    print(f"| {cls} | {cnt:,} | {pct:.2f}% |")

print(f"\n총 어노테이션 수: {total_annotations:,}")

print("\n#### 해상도 분포\n")
print("| 해상도      | 이미지 수 | 비율    |")
print("|---         |---:      |---:     |")
for res, cnt in res_cnt.items():
    pct = cnt / total_images * 100
    print(f"| {res} | {cnt:,} | {pct:.2f}% |")
```

```python
import os, json
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt

DATA_DIR = r"c:\\Users\\main\\Desktop\\val_labeled"
json_files = [f for f in os.listdir(DATA_DIR) if f.endswith(".json")]

class_cnt = Counter()
res_cnt = Counter()
area_list = []  # polygon 넓이 모음
objs_per_im = []  # 이미지당 객체 수
mixed_cnt = 0

IMG_W, IMG_H = 1280, 720  # 전부 동일-해상도
IMG_AREA = IMG_W * IMG_H

for fname in json_files:
    with open(os.path.join(DATA_DIR, fname), encoding="utf-8") as fp:
        obj = json.load(fp)

    # (1) 해상도 집계
    w = obj["images"]["image_width"]
    h = obj["images"]["image_height"]
    res_cnt[f"{w}×{h}"] += 1

    # (2) 어노테이션 집계 + polygon area 모으기
    anns = obj["annotations"]
    objs_per_im.append(len(anns))

    for ann in anns:
        cname = ann["class_name"]
        class_cnt[cname] += 1
        if cname == "mixed":
            mixed_cnt += 1
            print("mixed 샘플:", fname)

        # region 값 = polygon 넓이(px²)
        area_list.append(float(ann["region"]))

print(f"\n총 이미지: {len(json_files):,}")
print(f"총 어노테이션: {sum(class_cnt.values()):,}")
print(f"mixed 갯수 = {mixed_cnt}\n")

# ───────────────────────────────────────────────
# 1) polygon area(객체 크기) 분포
# ───────────────────────────────────────────────
areas = np.asarray(area_list)
small_thresh = 0.01 * IMG_AREA  # 이미지의 1% (= 9,216px²) 미만이면 small 로 정의

print("polygon area 통계(px²)")
for p in [0, 5, 25, 50, 75, 95, 100]:
    print(f"  {p:>2}퍼센타일 : {np.percentile(areas, p):8.0f}")

small_ratio = (areas < small_thresh).mean() * 100
print(f"\n작은 객체( <1% area ) 비율 : {small_ratio:.2f}%")

plt.figure()
plt.hist(areas, bins=50, log=True)
plt.axvline(small_thresh, color="r", linestyle="--", label="1% area")
plt.title("Polygon Area Distribution (log scale)")
plt.xlabel("area (pixel²)")
plt.ylabel("object count")
plt.legend()
plt.tight_layout()
plt.show()
```
